```python
import os
from glob import glob
import numpy  as np
import gdal
import earthpy.spatial as es
import earthpy.plot as ep
import rasterio
import sys
import matplotlib.pyplot as plt
```

```python
def imMap(date, ndvi):
    ep.plot_bands(ndvi, 
        cmap='Spectral',
        scale=False,
        title=date)
    plt.show()
```

```python
path = os.getcwd()+"\\working_clusters\\"
dists = glob(path+"*")
dists
```

```python
for dist in dists:
    imgs = glob(dist+"\\*.tif")
    print(imgs)
    for img in imgs:
        ds = gdal.Open(img)
        ndvi = np.array(ds.GetRasterBand(1).ReadAsArray()).astype(float)
        imMap(img.split('\\')[-1].split('.')[0], ndvi)
```
