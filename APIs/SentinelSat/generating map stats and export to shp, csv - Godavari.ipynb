```python
cwd = os.getcwd()
```

```python
def calculateArea(clusters):
    dense = np.count_nonzero(clusters == -1)
    sparse = np.count_nonzero(clusters == 1)
    return [sparse/100, dense/100]
```

```python
godavariFile = cwd + "\\GodavariMapShape\\GodavariMapShape.shp"
godavariShape = geopandas.read_file(godavariFile)
```

```python
ndviFolder = cwd + "\\working_ndvi\\Godavari"
clusterFolder = cwd + "\\working_clusters\\Godavari"
ndvi = glob(ndviFolder + "\\*.tif")
clusters = glob(clusterFolder + "\\*.tif")
```

```python
raster = rasterio.open(ndvi[0])
proj = raster.crs.to_proj4()
godavariShape = godavariShape.to_crs(proj)
godavariShape = godavariShape.dissolve(by=list(godavariShape)[6])
```

```python
godavariShape
```

```python
clusterStats = {}
for date in clusters:
    print(date.split('\\')[-1].split('.')[0])
    stats = zonal_stats(godavariShape, date, nodata = np.nan, stats="count", add_stats={'sparse_dense':calculateArea})
    print(stats)
    clusterStats[date.split('\\')[-1].split('.')[0]] = stats
```

```python
ndviStats = {}
for date in ndvi:
    print(date.split('\\')[-1].split('.')[0])
    stats = zonal_stats(godavariShape, date, nodata = np.nan, stats="min max mean std sum count")
    print(stats)
    ndviStats[date.split('\\')[-1].split('.')[0]] = stats
```

```python
shp_path = cwd + "\\GodavariStats"
if not os.path.exists(shp_path):
    os.mkdir(shp_path)
os.chdir(shp_path)
print(os.getcwd())
```

```python
regions = list(godavariShape.index.values)
```

```python
for i in range(len(clusters)):
    date = clusters[i].split('\\')[-1].split('.')[0]
    sparse = []
    dense = []
    means = []
    print(date)
    for region in clusterStats[date]:
        sparse.append(region["sparse_dense"][0])
        dense.append(region["sparse_dense"][1])
    for region in ndviStats[date]:
        means.append(region["mean"])
    fileName = date + '.csv'
    with open(fileName, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["AC_NAME", "Sparse", "Dense", "Mean NDVI"])
        for i in range(len(regions)):
            writer.writerow([regions[i], sparse[i], dense[i], means[i]])
```

```python
for i in range(len(clusters)):
    date = clusters[i].split('\\')[-1].split('.')[0]
    sparse = []
    dense = []
    means = []
    print(date)
    for region in clusterStats[date]:
        sparse.append(region["sparse_dense"][0])
        dense.append(region["sparse_dense"][1])
    for region in ndviStats[date]:
        means.append(region["mean"])
    dateShape = godavariShape.assign(sparseArea = sparse)
    dateShape['denseArea'] = dense
    dateShape['meanNDVI'] = means
    dateShape.to_file(date)
```
