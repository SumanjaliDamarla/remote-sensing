```python
ndvi_images = glob(path + "\\*.tif")
```

```python
for scene in ndvi_images:
    print(scene)
```

```python
def imMap(date, img, col):
    ep.plot_bands(img, 
        cmap=col,
        scale=False,
        title=date)
    plt.show()
```

```python
scenes = [ndvi_images[1], ndvi_images[2], ndvi_images[3], ndvi_images[7]]
cols = ["Blues", "Greens", "Reds", "Greys"]
band = ["Blue", "Green", "Red", "NIR"]
for i in range(len(scenes)):
    with rasterio.open(scenes[i]) as src:
        layer = src.read()
        imMap(band[i], layer, cols[i])
```
